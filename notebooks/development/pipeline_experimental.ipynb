```python
import omnipath as op
import pandas as pd
import mudata as mu
import networkx as nx

# 1. Konfiguracja globalna i pobranie danych z OmniPath
# Używamy oficjalnych, istniejących struktur API w Pythonie
MOUSE_TAX_ID = 'mouse'

print("Pobieranie sieci międzykomórkowej (Ligand-Receptor)...")
# Ta metoda pobiera zewnatrzkomórkowe pary L-R
# intercell_df = op.interactions.import_intercell_network(organism=MOUSE_TAX_ID)
intercell_df = op.interactions.import_intercell_network()

print("Pobieranie uniwersalnej sieci sygnalizacji wewnątrzkomórkowej...")
# AllInteractions.get() to oficjalna metoda pobierania monolitycznego grafu wewnątrzkomórkowego
global_signaling_df = op.interactions.AllInteractions.get(organism=MOUSE_TAX_ID)

print(f"Pobrane interkomórkowe: {intercell_df.shape}")
print(f"Pobrane wewnątrzkomórkowe: {global_signaling_df.shape}")

# 2. Dynamiczna ekstrakcja cech z obiektu MuData (RNA i MSI)
available_modalities = list(mdata.mod.keys())
rna_features = set(mdata['rna'].var_names) if 'rna' in available_modalities else set()
msi_features = set(mdata['msi'].var_names) if 'msi' in available_modalities else set()

print(f"\nWykryte modalności: {available_modalities}")
print(f"Cechy RNA: {len(rna_features)} | Cechy MSI (Metabolity): {len(msi_features)}")

# 3. Budowa zintegrowanego grafu sieciowego w NetworkX
integrated_graph = nx.DiGraph()

# STRATEGIA 1: Filtrowanie i integracja warstwy RNA (Wewnątrzkomórkowa sygnalizacja z AllInteractions)
if 'rna' in available_modalities:
    # Wybieramy tylko krawędzie skierowane, gdzie źródło i cel są w naszych danych RNA
    rna_directed = global_signaling_df[global_signaling_df["is_directed"] == True]
    
    filtered_rna = rna_directed[
        rna_directed["source_genesymbol"].isin(rna_features) & 
        rna_directed["target_genesymbol"].isin(rna_features)
    ]
    
    for _, row in filtered_rna.iterrows():
        integrated_graph.add_edge(
            row["source_genesymbol"],
            row["target_genesymbol"],
            is_stimulation=row["is_stimulation"],
            is_inhibition=row["is_inhibition"],
            layer="intracellular_rna"
        )
    print(f"Dodano krawędzie wewnątrzkomórkowe RNA: {filtered_rna.shape[0]}")

# STRATEGIA 2: Filtrowanie i integracja warstwy MSI (Metabolity z import_intercell_network)
if 'msi' in available_modalities:
    # Szukamy powiązań małych cząsteczek z MSI z receptorami RNA obecnymi w sieci intercell
    # Sprawdzamy kolumny identyfikujące typ encji (protein vs small_molecule)
    metabolite_src = intercell_df[
        (intercell_df["entity_type_source"] == "small_molecule") & 
        (intercell_df["source_genesymbol"].isin(msi_features)) &
        (intercell_df["target_genesymbol"].isin(rna_features))
    ]
    
    metabolite_tgt = intercell_df[
        (intercell_df["entity_type_target"] == "small_molecule") & 
        (intercell_df["target_genesymbol"].isin(msi_features)) &
        (intercell_df["source_genesymbol"].isin(rna_features))
    ]
    
    filtered_msi = pd.concat([metabolite_src, metabolite_tgt]).drop_duplicates()
    
    for _, row in filtered_msi.iterrows():
        # W sieci intercell sprawdzamy stymulację/inhibicję na podstawie dostępnych flag
        integrated_graph.add_edge(
            row["source_genesymbol"],
            row["target_genesymbol"],
            is_stimulation=row.get("is_stimulation", 1), 
            is_inhibition=row.get("is_inhibition", 0),
            layer="extracellular_msi"
        )
    print(f"Dodano krawędzie metaboliczne MSI: {filtered_msi.shape[0]}")

# 4. Podsumowanie struktury grafu pod kątem translacji do MaBoSS
print("\n--- PODSUMOWANIE GRAFU OMNIPHYSIBOSS ---")
print(f"Liczba węzłów w sieci boolowskiej: {integrated_graph.number_of_nodes()}")
print(f"Liczba krawędzi regulacyjnych: {integrated_graph.number_of_edges()}")

# Wyznaczenie węzłów wejściowych i wyjściowych do weryfikacji spójności szlaków
inputs = [n for n in integrated_graph.nodes if integrated_graph.in_degree(n) == 0]
outputs = [n for n in integrated_graph.nodes if integrated_graph.out_degree(n) == 0]

print(f"Liczba wejść (Input nodes): {len(inputs)} (np. {inputs[:5]})")
print(f"Liczba wyjść (Output nodes): {len(outputs)} (np. {outputs[:5]})")
```
