### Notatka - ustawienia modeli:

#### Ustawienia agent 
- Wewnętrzna sieć logiczna komórki (MaBoSS): Każdy agent (komórka) może posiadać własną instancję sieci logicznej. Możesz definiować dowolne przejścia stochastyczne, stałe szybkości aktywacji ($\alpha_i$) i inhibicji ($\beta_i$) dla poszczególnych węzłów.
- Mapowanie sygnałów wejściowych (Inputs): Możesz powiązać stężenie substancji w środowisku zewnętrznym (np. tlen, glukoza, TNF zaimportowane z warstwy dyfuzyjnej PhysiCell BioFVM) bezpośrednio ze stanem logicznym węzłów wejściowych sieci (np. węzeł_TNF = 1 gdy stężenie $> \text{próg}$).
- Mapowanie sygnałów wyjściowych (Outputs): Stan logiczny węzłów wyjściowych sieci (np. Apoptosis = 1, Proliferation = 1) steruje bezpośrednio flagami fenotypowymi agenta w PhysiCell (np. uruchomienie fazy śmierci komórkowej, przejście do fazy podziału G1/S).
- Heterogeniczność populacji: Każdy typ komórki (cell_definition) może ładować inne pliki .bnd i .cfg, co pozwala na symulację interakcji np. komórek nowotworowych z komórkami układu odpornościowego posiadającymi odmienne szlaki sygnałowe.
